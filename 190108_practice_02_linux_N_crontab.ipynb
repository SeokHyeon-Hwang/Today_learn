### 02 MYSQL 에서 데이터 가져오기<br>

#### 02-01 python에서 MYSQL DB에 session 연결<br>

```python
import mysql.connector

mydb = mysql.connector.connect(
    host = 'localhost',
    user = 'root',
    passwd = 'qwer1234',
    database = 'mydatabase'
)

mycursor = mydb.cursor()
```

#### 02-02 테이블 확인, 테이블 내 데이터 가죠오기<br>
##### 테이블 확인

```python
sql = 'show tables'
mycursor.execute(sql)
for x in mycursor:
    print(x)
```

##### 테이블 중 지마켓 테이블 내 데이터 가져오기

```python
sql = 'select*from gmarket'
mycursor.execute(sql)

for x in mycursor:
    print(x)
```
